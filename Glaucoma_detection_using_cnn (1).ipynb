```python
import os
import shutil

# Create the main combine directory if it doesn't exist
combine_dir = '/content/drive/MyDrive/combine'
if not os.path.exists(combine_dir):
    os.mkdir(combine_dir)

def combine_images(paths, dest):
    # Create the destination directory if it doesn't exist
    if not os.path.exists(dest):
        os.mkdir(dest)
    for path in paths:
        if os.path.exists(path) and os.listdir(path):
            shutil.copytree(path, dest, dirs_exist_ok=True)
        else:
            print(f"Warning: {path} does not exist or is empty.")
    print(f"Total images in {dest}: {len(os.listdir(dest))}")

# Define paths for glaucoma images
g_list = [
    '/content/drive/MyDrive/Acrima Dataset/Acrima Dataset/test/Glaucoma',
    '/content/drive/MyDrive/Acrima Dataset/Acrima Dataset/train/Glaucoma',
    '/content/drive/MyDrive/Rim One dataset/Rim One Dataset/RIM-ONE_DL_images/partitioned_randomly/test_set/glaucoma',
    '/content/drive/MyDrive/Rim One dataset/Rim One Dataset/RIM-ONE_DL_images/partitioned_randomly/training_set/glaucoma',
    '/content/drive/MyDrive/drishtiimage/archive/Test-20211018T060000Z-001/Test/Images/glaucoma',
    '/content/drive/MyDrive/drishtiimage/archive/Training-20211018T055246Z-001/Training/Images/GLAUCOMA'
]
g_dest = os.path.join(combine_dir, 'glaucoma')

# Combine glaucoma images
combine_images(g_list, g_dest)

# Define paths for normal images
n_list = [
    '/content/drive/MyDrive/Acrima Dataset/Acrima Dataset/test/Non Glaucoma',
    '/content/drive/MyDrive/Acrima Dataset/Acrima Dataset/train/Non Glaucoma',
    '/content/drive/MyDrive/Rim One dataset/Rim One Dataset/RIM-ONE_DL_images/partitioned_randomly/test_set/normal',
    '/content/drive/MyDrive/Rim One dataset/Rim One Dataset/RIM-ONE_DL_images/partitioned_randomly/training_set/normal',
    '/content/drive/MyDrive/drishtiimage/archive/Test-20211018T060000Z-001/Test/Images/normal',
    '/content/drive/MyDrive/drishtiimage/archive/Training-20211018T055246Z-001/Training/Images/NORMAL'
]
n_dest = os.path.join(combine_dir, 'normal')

# Combine normal images
combine_images(n_list, n_dest)
```

```python

img_height=256
img_width=256
batch_size=32
datagen = ImageDataGenerator(rescale=1./255,
shear_range = 0.15,
zoom_range = 0.15,
horizontal_flip = True)
train_ds = datagen.flow_from_directory(
    '/content/images/train',
    target_size = (img_height, img_width),
    batch_size = batch_size,
    class_mode='categorical',
    shuffle=False)

val_ds = datagen.flow_from_directory(
    '/content/images/val',
    target_size = (img_height, img_width),
    batch_size = batch_size,
    class_mode='categorical',
    shuffle=False)

test_ds = datagen.flow_from_directory(
    '/content/images/test',
    target_size = (img_height, img_width),
    batch_size = batch_size,
    class_mode='categorical',
    shuffle=False)


print("Training directory contents:", os.listdir('images/train'))
print("Validation directory contents:", os.listdir('images/val'))
print("Test directory contents:", os.listdir('images/test'))
#print(val_ds)
print(len(test_ds))
print(len(train_ds))
print(train_ds)
print("Value of val_ds:", len(val_ds))
```

```python
!pip install -q tensorflow-model-optimization  # Install the missing module
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import Input
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense,Dropout
from keras.layers import BatchNormalization
import tf_keras
from tf_keras import layers
import tensorflow as tf
import tensorflow_model_optimization as tfmot


# Initialising the CNN
classifier = Sequential()
#input_layer = Input(shape=(256, 256, 3))
# Step 1 - Adding Convolution layer
classifier.add(Conv2D(32, (3, 3), input_shape=(256, 256, 3), activation = 'relu'))

# Step 2 - Adding MaxPooling layers
classifier.add(MaxPooling2D(pool_size = (2, 2)))
# Adding a second convolutional layer
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Step 3 - Flattening
classifier.add(Flatten())

# Step 4 - Full connection
classifier.add(Dense(units = 512, activation = 'relu'))
classifier.add(BatchNormalization()),
classifier.add(Dense(256,activation='relu')),
classifier.add(Dropout(0.25)),
classifier.add(Dense(units = 3, activation = 'softmax'))

# Compiling the CNN
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python

model_info=classifier.fit(train_ds,
epochs = 120,
validation_data = val_ds,
verbose = 1)
```
